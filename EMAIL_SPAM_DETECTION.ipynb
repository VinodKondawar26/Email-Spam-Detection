```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
maildata=pd.read_csv("/content/spam.csv",encoding="latin-1")
maildata
```

```python
maildata.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis=1,inplace=True)
maildata
```

```python
maildata["class"]=maildata["class"].map({'ham':0,'spam':1})
maildata
```

```python
cv=CountVectorizer()
x=maildata["message"]
y=maildata["class"]
(x.shape,y.shape)
```

```python
x=cv.fit_transform(x)

x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=MultinomialNB()
model.fit(x_train,y_train)
model.score(x_test,y_test)*100
```

```python
import pickle
pickle.dump(model,open("maildata.pkl","wb"))
```

```python
pickle.dump(cv,open("vectorizer.pkl","wb"))
```
